```python
import pandas as pd
reviews = pd.read_csv("winemag-data-130k-v2.csv", index_col=0)

```

# Exercise 1 – Data type of the column

```python
dtype = reviews.points.dtype
dtype
```

# Exercise 2 – Converts column to strings

```python
point_strings = reviews.points.astype(str)
point_strings
```

# Exercise 3 – Counts missing prices

```python
n_missing_prices = reviews.price.isnull().sum()
n_missing_prices
```

# Exercise 4 – Most common wine‑producing regions

```python
reviews_per_region = (
    reviews.region_1
        .fillna('Unknown')
        .value_counts()
        .sort_values(ascending=False)
)

reviews_per_region
```
